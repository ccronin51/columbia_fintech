For this categorical columns, I will try to find the number of unique classes in every column to decide the `encoding type`

```python
# Filter non-numerical (categorical) columns
categorical_columns = loans_df.select_dtypes(exclude=[np.number]).columns

# Initialize an empty list to store results
unique_counts = []

# Iterate through each categorical column to calculate the number of unique classes
for column in categorical_columns:
    unique_count = loans_df[column].nunique()
    percentage = (unique_count / len(loans_df)) * 100
    unique_counts.append({'Column': column, 'Unique Count': unique_count, 'Percentage': percentage})

# Convert list to DataFrame
unique_counts_df = pd.DataFrame(unique_counts)

# Display the results
unique_counts_df
```

```python
# unique_counts_df.to_csv('unique_counts.csv', index=False)

def suggest_encoding(unique_counts):
    encoding_suggestions = {}
    for index, row in unique_counts.iterrows():
        column_name = row['Column']
        unique_value_count = row['Unique Count']
        
        if unique_value_count > 100:  # Adjust thresholds as necessary
            encoding_type = 'Target Encoding'
        elif unique_value_count > 2:
            encoding_type = 'Ordinal Encoding'
        else:
            encoding_type = 'Label Encoding'
        
        encoding_suggestions[column_name] = encoding_type
    
    return encoding_suggestions

# Example usage:
encoding_suggestions = suggest_encoding(unique_counts_df)
encoding_suggestions
```

```python
# Initialize lists to hold the column names for each encoding type
target_encoding_columns = []
ordinal_encoding_columns = []
label_encoding_columns = []

# Loop through the dictionary and append column names to the respective lists based on encoding type
for column, encoding_type in encoding_suggestions.items():
    if encoding_type == 'Target Encoding':
        target_encoding_columns.append(column)
    elif encoding_type == 'Ordinal Encoding':
        ordinal_encoding_columns.append(column)
    elif encoding_type == 'Label Encoding':
        label_encoding_columns.append(column)

# Display the lists
print("Target Encoding Columns:", target_encoding_columns)
print("Ordinal Encoding Columns:", ordinal_encoding_columns)
print("Label Encoding Columns:", label_encoding_columns)
```

```python
import pandas as pd
import lightgbm as lgb
from sklearn.model_selection import train_test_split
from category_encoders import TargetEncoder, OrdinalEncoder
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, roc_auc_score, roc_curve



# Separate features and target
X = loans_df.drop('risk_flag', axis=1)
y = loans_df['risk_flag']


# Specify parameters explicitly to avoid FutureWarnings in TargetEncoder
target_encoder = TargetEncoder(cols=target_encoding_columns, min_samples_leaf=1, smoothing=1.0)
ordinal_encoder = OrdinalEncoder(cols=ordinal_encoding_columns)

# Apply Target and Ordinal Encoding
X_encoded = target_encoder.fit_transform(X, y)
X_encoded = ordinal_encoder.fit_transform(X_encoded, y)

# Apply Label Encoding individually for each column
label_encoder = {}
for col in label_encoding_columns:
    le = LabelEncoder()
    X_encoded[col] = le.fit_transform(X_encoded[col].astype(str))
    label_encoder[col] = le  # Store encoder if needed for inverse_transform or similar

for col in X_encoded.columns:
    X_encoded[col] = X_encoded[col].fillna(-1)

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)

# Initialize and train the LightGBM model
model = lgb.LGBMClassifier(force_col_wise=True, random_state=42)
model.fit(X_train, y_train)

# Predict probabilities for AUC computation
y_scores = model.predict_proba(X_test)[:, 1]  # get probabilities for the positive class

# Calculate AUC
auc_score = roc_auc_score(y_test, y_scores)
print("AUC Score:", auc_score)

# Compute ROC curve
fpr, tpr, _ = roc_curve(y_test, y_scores)

# Plot ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % auc_score)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC)')
plt.legend(loc="lower right")
plt.show()
```

For more realistic modelling, I will add class weight into consideration 

```python
from sklearn.utils.class_weight import compute_class_weight
import numpy as np

# Compute class weights if you want to specify them manually
class_weights = compute_class_weight('balanced', classes=np.unique(y_train), y=y_train)
weight_dict = dict(enumerate(class_weights))
weight_dict
```

```python
# Initialize and train the LightGBM model with class weights
model = lgb.LGBMClassifier(random_state=42, class_weight=weight_dict)  
model.fit(X_train, y_train)

# Predict probabilities for AUC computation
y_scores = model.predict_proba(X_test)[:, 1]

# Calculate AUC
auc_score = roc_auc_score(y_test, y_scores)
print("AUC Score:", auc_score)

# Compute ROC curve
fpr, tpr, _ = roc_curve(y_test, y_scores)

# Plot ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % auc_score)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC)')
plt.legend(loc="lower right")
plt.show()
```
