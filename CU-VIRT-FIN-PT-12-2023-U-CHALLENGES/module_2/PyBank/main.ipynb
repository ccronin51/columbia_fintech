```python
# Imports
from pathlib import Path
import csv
import sys
```

```python
# Set the file path
path = Path('./Resources/budget_data.csv')

# opening path as obj
with open(path,'r') as budget_data_obj:
    
    # output file handle
    beg_stdout = sys.stdout
    
    # opening blank text file write code to
    with open('financial_analysis.txt', 'w') as f:
        
        # reading in the csv file without header
        sys.stdout = f
        no_header = next(budget_data_obj)
        budget_data_reader = csv.reader(budget_data_obj)

        # initiatizing variables
        counter = 0
        total_profit = 0 
        total_months = 0
        months = []
        pnl = []
        max_month_chg = ['aaa', 0]
        min_month_chg = ['aaa', 0]
        
        # looping through budget data
        for row in budget_data_reader:

            counter += 1

            # adding months to months list
            if row[0] not in months:
                months.append(row[0])
            
            # adding profit and losses to pnl list
            pnl.append(int(row[1]))
            
            # caclulating total months and profit
            total_months += 1
            total_profit += int(row[1])

        # creating list of changes in profit from month to month, then calc average    
        pnl_changes = [y - x for x, y in zip(pnl, pnl[1:])]
        avg_changes = int(sum(pnl_changes) / len(pnl_changes))

        # creating tuple of months along with changes in profit from pnl_changes
        pnl_changes_w_month = [(x, y) for x,y in zip(months[1:], pnl_changes)]

        # looping through list of tuples to find max and min changes in profit with corresponding month
        for i,j in pnl_changes_w_month:
            if int(j) > 0 and int(j) > int(max_month_chg[1]):
                max_month_chg[0] = i
                max_month_chg[1] = j

            if int(j) < 0 and int(j) < int(min_month_chg[1]):
                min_month_chg[0] = i
                min_month_chg[1] = j

        print('Financial Analysis')
        print(' ')
        print('....................')
        print(' ')
        print('Total Months: ', total_months)
        print(' ')
        print('Total PnL: ' "${:,}".format(total_profit))
        print(' ')
        print('Avgerage Change in Profit: ' "${:,.0f}".format(avg_changes))
        print(' ')
        print('Greatest Increase in Profits: ', max_month_chg[0], " ", "${:,}".format(int(max_month_chg[1])))
        print(' ')
        print('Greatest Decrease in Profits: ', min_month_chg[0], " ", "${:,}".format(int(min_month_chg[1])))

        sys.stdout = beg_stdout 
    
    # Printing Results to Command Line
    print('Financial Analysis')
    print(' ')
    print('....................')
    print(' ')
    print('Total Months: ', total_months)
    print(' ')
    print('Total PnL: ' "${:,}".format(total_profit))
    print(' ')
    print('Avgerage Change in Profit: ' "${:,.0f}".format(avg_changes))
    print(' ')
    print('Greatest Increase in Profits: ', max_month_chg[0], " ", "${:,}".format(int(max_month_chg[1])))
    print(' ')
    print('Greatest Decrease in Profits: ', min_month_chg[0], " ", "${:,}".format(int(min_month_chg[1])))
```
